```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns
sns.set()
```

```python
metrics = pd.read_csv('logs/PPO_v0_1400.csv')
```

```python
metrics
```

```python
plt.plot(metrics.episode, metrics.test_win_rate.rolling(1).mean())
plt.plot(metrics.episode, metrics.train_win_rate.rolling(1).mean())
plt.legend(['Rolling test winrate', 'Rolling train winrate'])
# plt.title('Rolling test winrate')
plt.xlabel('episode')
plt.show()
```

```python
plt.plot(metrics.episode, metrics.total_score)
plt.title('Score')
plt.xlabel('episode')
plt.show()
```

```python
plt.plot(metrics.episode, metrics.train_loss_policy)
plt.plot(metrics.episode, metrics.train_loss_value)
plt.title('Losses')
plt.legend(['policy', 'value'])
plt.xlabel('episode')
plt.show()
```
